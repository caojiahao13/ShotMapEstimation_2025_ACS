# Simulation

We implement 
* The proposed Jointly Spatial-Varying Log-Gaussian Cox Process Model (denoted by `JSVLGCP`)

alongside several competitive methods 

* standard Log-Gaussian Cox Process model (denoted by `LGCP`),
* nonhomogeneous Poisson process model (denoted by `IPP`),
* kernel density estimation (denoted by `KDE`),
* Bayesian Additive Regression Trees (denoted by `BART`).

The fitting process can take hours, but the results are already stored. You can skip the model fitting and proceed directly to the visualization steps to generate Figures 3 and 4 for the manuscript.

```r
library(RandomFields)
library(fields)
library(orthopolynom)
library(mvtnorm)
library(combinat)
library(lattice)
library(Matrix)
library(limSolve)
library(MCMCpack)
library(truncnorm)
library(ggplot2)
library(tidyverse)
library(doParallel)
library(foreach)
library(spatstat)
library(inlabru)
library(sf)
library(sp)
library(INLA)
library(BART)
library(splines)
source("../RealDataAnalysis/Functions/basket_func_same.R")
source("../RealDataAnalysis/Functions/SVC-LGCP.R")

save_folder = 'Data'
if (!dir.exists(save_folder)) {
      dir.create(save_folder)
}
simu_info = list(
    num_rep = 20,
    num_cores = 10, 
    settings = c('simu','curry'),
    num_save = 1000,
    num_burn = 4000,
    update_saved = FALSE
)

simu_setting_df <- expand.grid(setting = simu_info$settings, i_rep = seq(simu_info$num_rep))
num_simu_setting = dim(simu_setting_df)[1]
```

```r
cl = makeCluster(simu_info$num_cores, outfile=sprintf('%s/dopar_log.out', save_folder))
registerDoParallel(cl)

result__ = foreach(i_simu_setting = 1:num_simu_setting, 
              .packages = c("RandomFields","orthopolynom","mvtnorm","combinat","lattice","Matrix",
                           "limSolve","MCMCpack","truncnorm","spatstat","splines","inlabru","sf","sp","INLA","fields","BART","splines","ggplot2"), .verbose = TRUE, .errorhandling = "stop") %dopar% {
    source("../RealDataAnalysis/Functions/basket_func_same.R")
    source("../RealDataAnalysis/Functions/SVC-LGCP.R")

    # Load setting
    # -------------------------------------
    i_rep = simu_setting_df$i_rep[i_simu_setting]
    setting = simu_setting_df$setting[i_simu_setting]
    
    # Create subfolder for settings
    # -------------------------------------
    setting_string = setting
    save_subfolder = sprintf('%s/%s', save_folder, setting_string)
    if (!dir.exists(save_subfolder)) {
      dir.create(save_subfolder)
    }
 
    # Define mesh
    # -------------------------------------
    boundary = fm_segm(rbind(c(0, 0), c(1, 0), c(1, 1), c(0, 1))*2-1, is.bnd = TRUE, crs = NA_character_)
    delta_lattice = 0.05
    lattice = fm_lattice_2d(x = seq(-1,1,by = delta_lattice), y = seq(-1,1,by = delta_lattice), type = 'R', boundary = boundary)
    mesh = fm_rcdt_2d(lattice = lattice, boundary = boundary, crs = NA_character_) # ~120K vertex
    vrt = fm_vertices(mesh) # spdf for vertex coordinates
    vrt_coords = st_coordinates(vrt)[,1:2] # matrix with columns X and Y

    boundary_poly = Polygon(rbind(c(0, 0), c(1, 0), c(1, 1), c(0, 1))*2-1)
    boundary_poly = SpatialPolygons(list(Polygons(list(boundary_poly), ID = '1')))

    window = owin(c(-1,1),c(-1,1))



    
    # # --------------------# --------------------# --------------------
    # # 1. DGP
    # # --------------------# --------------------# --------------------
    if(setting == 'simu'){
        num_game = 200
        covar = data.frame(game = seq(num_game), home = as.numeric(seq(num_game) %% 2 == 1), 
                           away = as.numeric(seq(num_game) %% 2 == 0), level = as.numeric(seq(num_game) %% 4 <= 1))
        # 1.1 define intensity function
        # --------------------
        p=dim(covar)[2]-1
        courtgrid = as.matrix(expand.grid(seq(-1,1,length=50),seq(-1,1,length=40)))
        intergrid = as.matrix(expand.grid(seq(-1,1,length=60),seq(-1,1,length=60)))
        dxdy=(intergrid[1,1]-intergrid[2,1])^2
        a = 1
        d = 2
        b = 1
        threshold = 0.7
        # GP.eigen.degree(threshold,a=a,b=b,d=d)
        n = GP.eigen.degree(threshold,a=a,b=b,d=d)$n
        eigmat = GP.eigen.value(n=n,a=a,b=b,d=d)
        eigmat = diag(eigmat)
        pmat = eigmat
        L = nrow(eigmat)
    }
    if(setting == 'curry'){
        num_game = 200
        covar = data.frame(game = seq(num_game), home = as.numeric(seq(num_game) %% 2 == 1), 
                           away = as.numeric(seq(num_game) %% 2 == 0), level = as.numeric(seq(num_game) %% 4 <= 1))
        # 1.1 define intensity function
        # --------------------
        p=dim(covar)[2]-1
        courtgrid = as.matrix(expand.grid(seq(-1,1,length=50),seq(-1,1,length=40)))
        intergrid = as.matrix(expand.grid(seq(-1,1,length=60),seq(-1,1,length=60)))
        dxdy=(intergrid[1,1]-intergrid[2,1])^2
        a = 0.25
        d = 2
        b = 1.5
        threshold = 0.8
        #GP.eigen.degree(0.9,a,b=9,d=2)
        # GP.eigen.degree(0.8,a=a,b=b,d=d)
        n = GP.eigen.degree(threshold,a=a,b=b,d=d)$n
        eigmat = GP.eigen.value(n=n,a=a,b=b,d=d)
        eigmat = diag(eigmat)
        pmat = eigmat
        L = nrow(eigmat)
    }

    intermat = GP.eigen.funcs(intergrid, n=n, a=a, b=b)
    courtmat = GP.eigen.funcs(courtgrid, n=n, a=a, b=b)
    num.type = 2

    if(setting == 'simu'){
        set.seed(1)    
        theta = 2*matrix(runif(L*(1+num.type*p))-0.5, L, 1+num.type*p) # 4 for 2426
    }
    if(setting == 'curry'){
        load("../RealDataAnalysis/Data/curry/curry_homelevel_theta.RData")
        theta = apply(Theta.result[10001:15000,,], c(2,3), mean)
    }
    theta_alpha = theta[,1,drop = FALSE]
    theta_beta_type0 = theta[,2:(1+p),drop = FALSE]
    theta_beta_type1 = theta[,(2+p):(1+2*p),drop = FALSE]
    
    logintensity_function <- function(x, y, z, type, theta_alpha, theta_beta_type0, theta_beta_type1, n, a, b) {

        coords = matrix(0, length(x), 2)
        coords[,1] = x;  coords[,2] = y 
        GP_basis_value = GP.eigen.funcs(coords, n=n, a=a, b=b) 

        if(type == 0) logintensity = GP_basis_value%*%theta_alpha + GP_basis_value %*% theta_beta_type0 %*% z
        if(type == 1) logintensity = GP_basis_value%*%theta_alpha + GP_basis_value %*% theta_beta_type1 %*% z

      return( as.vector(logintensity) ) 
    }
    intensity_function <- function(x, y, z, type , theta_alpha, theta_beta_type0, theta_beta_type1, n, a, b) {
        return(exp(logintensity_function(x = x, y = y, z = z, 
                                         type =type, theta_alpha = theta_alpha, theta_beta_type0 = theta_beta_type0, 
                                         theta_beta_type1 = theta_beta_type1, n = n, a = a, b = b)))
    }
    
    temp_intensity_range0 = range(intensity_function(x = intermat[,1], y = intermat[,2], z = rep(1,p), type = 0, theta_alpha, theta_beta_type0, theta_beta_type1, n, a, b))
    temp_intensity_range1 = range(intensity_function(x = intermat[,1], y = intermat[,2], z = rep(1,p), type = 1, theta_alpha, theta_beta_type0, theta_beta_type1, n, a, b))
    temp_intensity_range = c(temp_intensity_range0,temp_intensity_range1)
    print(temp_intensity_range)
    if(max(temp_intensity_range)/min(temp_intensity_range)>1000){
        print('Bad Setting!')
    }
    # 1.2 simulate from Poisson process: shot_data and covar
    # --------------------
    save_file_name = sprintf('%s/DATA_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){
        # Generate PP for miss shots for each game
        shot_data = data.frame(Game = 1, Missmade = 1, Xdist = 1, Ydist = 1)
        intensity_values = c()
        for(i_game in seq(num_game)){ 
            set.seed(i_rep*10000 + i_game)
            for(type in c(0,1)){
                z = as.numeric(covar[i_game, 2:(p + 1)])
                # shot_ppp = rpoispp(lambda = intensity_function, lmax=NULL, win=window, nsim = 1, 
                #                   z = z, type = type, 
                #                   theta_alpha = theta_alpha, theta_beta_type0 = theta_beta_type0, theta_beta_type1 = theta_beta_type1,
                #                   n = n, a = a, b = b)
                target.crs <- fm_CRS(NA_character_)  # weird bug
                shot_spdf = sample.lgcp(mesh, 
                                loglambda = 
                                    logintensity_function(x = vrt_coords[,1], y= vrt_coords[,2],
                                        z = z, type = type, 
                                        theta_alpha = theta_alpha, theta_beta_type0 = theta_beta_type0, theta_beta_type1 = theta_beta_type1,
                                        n = n, a = a, b = b),
                                strategy = 'rectangle', ignore.CRS = TRUE
                                )    # A spdf
                shot_coords = shot_spdf@coords
                if(dim(shot_coords)[1]>0){
                    shot_data_gamei = data.frame(Game = i_game, Missmade = type, Xdist = shot_coords[,1], Ydist = shot_coords[,2])
                    shot_data = rbind(shot_data, shot_data_gamei)
    
                    intensity_values_game_type = intensity_function(x = shot_coords[,1], y = shot_coords[,2],
                                                                      z = z, type = type, 
                                                                      theta_alpha = theta_alpha, theta_beta_type0 = theta_beta_type0, theta_beta_type1 = theta_beta_type1,
                                                                      n = n, a = a, b = b)
                    intensity_values = c(intensity_values, intensity_values_game_type)
                }
    
        }}
        shot_data = shot_data[-1, ]
        shot_data = cbind(shot_data, covar[shot_data$Game,c('home','away','level')])
        save(shot_data, intensity_values, covar, theta, n, a, b, threshold, file = save_file_name)
    }else{
        load(file = save_file_name)
    }
    shot_coords = as.matrix(shot_data[,c('Xdist','Ydist')])

    shot_data_sp = SpatialPoints(coords = shot_coords, bbox = as.matrix(data.frame(min = c(-1,-1), max = c(1,1))))
    shot_data_spdf = SpatialPointsDataFrame(coords = shot_coords, 
                                            data = cbind(shot_data$Missmade, covar[shot_data$Game,c('home','away','level')]), 
                                            bbox = as.matrix(data.frame(min = c(-1,-1), max = c(1,1))), match.ID = FALSE)

    shot_data_ppp = ppp(x = shot_data$Xdist, y = shot_data$Ydist, window = window )
    shot_data_ppp_mark = ppp(x = shot_data$Xdist, y = shot_data$Ydist, window = window, marks = covar[shot_data$Game,c('home','away','level')] )
        
    num_data = dim(shot_coords)[1]
    
    if(i_rep == 1){
        figure = ggplot(data = shot_data) + geom_point(aes(x = Xdist, y = Ydist, color = Missmade)) + facet_grid(rows = vars(home), cols = vars(level))
        ggsave(figure, file = sprintf('%s/Points_rep=%d.pdf', save_subfolder, i_rep))
    }

    # --------------------# --------------------# --------------------
    # 2. Fitting
    # --------------------# --------------------# --------------------
    # 2.1 SVC_LGCP
    # --------------------
    save_file_name = sprintf('%s/SVCLGCP_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){

        num.type = 2
        apop = 5
        bpop = 5
        sigma.pop = 1#rinvgamma(1,apop,bpop)
        atype = 5
        btype = 5
        sigma.type = rep(1,num.type*p)#rinvgamma(num.type*p,atype,btype)
        N = simu_info$num_burn + simu_info$num_save
        fit = fit_SVCLGCP(shot_data, covar) 
        
        ##: list(Theta.result = Theta.result, likelihood1 = likelihood1, Sigmapop = Sigmapop, Sigmatype = Sigmatype)
        ##: fit$Theta.result: [isim, i_L, j_type]

        intensity_pred = rep(0 , num_data)
        theta_hat = apply(fit$Theta.result[simu_info$num_burn+1:simu_info$num_save,,],c(2,3),mean)
        for(home in c(0,1)){for(level in c(0,1)){for(type in c(0,1)){
            id_in_home_level_type = which((covar[shot_data$Game, 'home'] == home) & (covar[shot_data$Game, 'level'] == level) & (shot_data$Missmade == type) )
            intensity_pred[id_in_home_level_type] = intensity_function(shot_coords[id_in_home_level_type, 1],  shot_coords[id_in_home_level_type, 2],
                                            c(home, 1 - home, level),
                                            type = type,
                                            theta_alpha = theta_hat[,1],
                                            theta_beta_type0 = theta_hat[,2:(p+1)],
                                            theta_beta_type1 = theta_hat[,(p+2):(2*p+1)],
                                            n = n, a = a, b = b)
    
        }}}
        save(fit, intensity_pred, file = save_file_name)
    }
    
   
    # 2.2 LGCP
    # --------------------
    save_file_name = sprintf('%s/LGCP_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){
        
        fit_list = vector(mode = 'list', length = 8)
        intensity_pred = c()
        counter = 0
        for(home in c(0,1)){for(level in c(0,1)){for(type in c(0,1)){ 
            counter = counter + 1
            num_game_home_level = sum( (covar[, 'home'] == home) & (covar[, 'level'] == level) )
            
            id_home_level_type = which((covar[shot_data$Game, 'home'] == home) & (covar[shot_data$Game, 'level'] == level) & (shot_data$Missmade == type) )
            
            # fit = kppm(shot_data_ppp[id_home_level_type, ], ~1 , "LGCP", covmodel=list(model="exponential"))
            # fit_list[[counter]] = fit                                   
            # intensity_pred[ id_home_level_type ] = predict(fit, locations = shot_data_ppp[id_home_level_type,]) / num_game_home_level
            matern = inla.spde2.pcmatern(mesh,
                    prior.sigma = c(5, 0.1),
                    prior.range = c(2, 0.1))
            fit = lgcp(components = coordinates ~ Intercept(1) + mySPDE(map = coordinates, model = matern), 
                        data = shot_data_spdf[id_home_level_type,],
                        domain = list(coordinates = mesh),
                        samplers = boundary_poly)
            fit_list[[counter]] = fit 
            intensity_pred[ id_home_level_type ] = predict(fit, shot_data_spdf[id_home_level_type,], ~ exp(mySPDE + Intercept))$mean / num_game_home_level
        }}}
        save(intensity_pred, file = save_file_name)
    }

    # 2.3 Inhomogeneous Poisson process
    # --------------------
    save_file_name = sprintf('%s/IPP_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){

        fit_list = vector(mode = 'list', length = 8)
        intensity_pred = c()
        counter = 0
        for(home in c(0,1)){for(level in c(0,1)){for(type in c(0,1)){ 
            counter = counter + 1
            num_game_home_level = sum( (covar[, 'home'] == home) & (covar[, 'level'] == level) )
            
            id_home_level_type = which((covar[shot_data$Game, 'home'] == home) & (covar[shot_data$Game, 'level'] == level) & (shot_data$Missmade == type) )
            
            fit = ppm(shot_data_ppp[id_home_level_type, ], ~ bs(x, 6) + bs(y, 6))
            fit_list[[counter]] = fit
            intensity_pred[ id_home_level_type ] = predict(fit, locations = shot_data_ppp[id_home_level_type,]) / num_game_home_level
        }}}
        save(fit_list, intensity_pred, file = save_file_name)

    }
    
    # 2.4 KDE
    # --------------------
    save_file_name = sprintf('%s/KDE_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){
        sigma_kde_list = c(seq(0.05, 1, by = 0.1), seq(1.2,2,by = 0.2))
        
        rmse = Inf
        sigma_kde_best = 0
        intensity_pred_best = 0
        
        for(sigma_kde in sigma_kde_list){
            intensity_pred = c()
            for(home in c(0,1)){for(level in c(0,1)){for(type in c(0,1)){ 
                num_game_home_level = sum( (covar[, 'home'] == home) & (covar[, 'level'] == level) )

                id_home_level_type = which((covar[shot_data$Game, 'home'] == home) & (covar[shot_data$Game, 'level'] == level) & (shot_data$Missmade == type) )

                fit = density.ppp(ppp(shot_coords[id_home_level_type,1], shot_coords[id_home_level_type,2], c(-1,1),c(-1,1)), sigma = sigma_kde)
                intensity_pred[id_home_level_type] = interp.im(fit, x = shot_coords[id_home_level_type,1], y = shot_coords[id_home_level_type,2])/num_game_home_level
            }}}
            rmse_new = sqrt(mean( intensity_pred - intensity_values )^2)  
            if(rmse_new < rmse){
                rmse = rmse_new
                sigma_kde_best = sigma_kde
                intensity_pred_best = intensity_pred
            }
        }
        
        save(sigma_kde_best, intensity_pred_best, file = save_file_name)
    }

    # 2.5 BART
    # --------------------
    save_file_name = sprintf('%s/BART_rep=%d.RData', save_subfolder, i_rep)
    if(!file.exists(save_file_name) | (simu_info$update_saved == TRUE) ){

        delta_lattice_temp = 0.05
        lattice_temp = fm_lattice_2d(x = seq(-1,1,by = delta_lattice_temp), y = seq(-1,1,by = delta_lattice_temp), type = 'R', boundary = boundary)
        lattice_coords_temp = lattice_temp$loc
        
        x.train = matrix(0, 1, 2 + 3)
        shot_count = c()
        for(home in c(0,1)){for(level in c(0,1)){for(type in c(0,1)){ 
            num_game_home_level = sum( (covar[, 'home'] == home) & (covar[, 'level'] == level) )
            
            id_home_level_type = which((covar[shot_data$Game, 'home'] == home) & (covar[shot_data$Game, 'level'] == level) & (shot_data$Missmade == type) )
            
            D_temp = rdist(shot_coords[id_home_level_type, ], lattice_coords_temp)
            shot_count_home_level_type = apply(D_temp, 2, function(d) return(sum(d<(delta_lattice_temp/2))))
                                               
            shot_count = c(shot_count, shot_count_home_level_type)
            x.train = rbind(x.train, cbind(lattice_coords_temp, matrix(c(home, level, type), nrow = dim(lattice_coords_temp)[1], ncol = 3, byrow = TRUE) ))
        }}}                                 
        x.train = x.train[-1, ]

        # fit = gbart(x.train = x.train, y.train = shot_count, ntree = 50, nskip = simu_info$num_burn, ndpost = simu_info$num_save)
        # intensity_pred = apply(predict( fit, newdata = cbind(shot_coords, shot_data_spdf@data[,-3]) ),2,mean) / (delta_lattice_temp^2) / num_game_home_level
                                               
        shot_count[shot_count > 1] = 1        
        fit = gbart(x.train = x.train, y.train = shot_count, ntree = 50, type = 'pbart', nskip = simu_info$num_burn, ndpost = simu_info$num_save)
        intensity_pred = predict( fit, newdata = cbind(shot_coords, shot_data_spdf@data[,-3]) )$prob.test.mean / (delta_lattice_temp^2) / num_game_home_level

        save(intensity_pred, file = save_file_name)   
    }


}

stopCluster(cl)
                                               
```

```r
library(RandomFields)
library(fields)
library(orthopolynom)
library(mvtnorm)
library(combinat)
library(lattice)
library(Matrix)
library(limSolve)
library(MCMCpack)
library(truncnorm)
library(ggplot2)
library(tidyverse)
library(doParallel)
library(foreach)
library(spatstat)
library(inlabru)
library(sf)
library(sp)
library(INLA)
library(BART)
library(splines)
library(ggbasketball)
library(gridExtra)
simu_info = list(
    num_rep = 20,
    num_cores = 20, 
    settings = c('simu','curry'),
    num_save = 1000,
    num_burn = 4000,
    update_saved = FALSE
)
save_folder = 'Data'

simu_setting_df <- expand.grid(setting = simu_info$settings, i_rep = seq(simu_info$num_rep))
num_simu_setting = dim(simu_setting_df)[1]

df_result_SVCLGCP = df_result_LGCP = df_result_IPP = df_result_KDE= df_result_BART = simu_setting_df

for(i_simu_setting in 1:num_simu_setting){

    # Load setting
    # -------------------------------------
    i_rep = simu_setting_df$i_rep[i_simu_setting]
    setting = simu_setting_df$setting[i_simu_setting]
    setting_string = setting
    save_subfolder = sprintf('%s/%s', save_folder, setting_string)
    
    # Load Data
    # -------------------------------------
    save_file_name = sprintf('%s/DATA_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    shot_coords = as.matrix(shot_data[,c('Xdist','Ydist')])
    num_data = dim(shot_coords)[1]
    
    
    # 2.1 SVC_LGCP
    # --------------------
    save_file_name = sprintf('%s/SVCLGCP_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    df_result_SVCLGCP$RMSE[i_simu_setting] = sqrt(mean((intensity_pred - intensity_values)^2))

    # 2.2 LGCP
    # --------------------
    save_file_name = sprintf('%s/LGCP_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    df_result_LGCP$RMSE[i_simu_setting] = sqrt(mean((intensity_pred - intensity_values)^2))
    
    # 2.3 IPP
    # --------------------
    save_file_name = sprintf('%s/IPP_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    df_result_IPP$RMSE[i_simu_setting] = sqrt(mean((intensity_pred - intensity_values)^2))
    
    # 2.4 KDE
    # --------------------
    save_file_name = sprintf('%s/KDE_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    df_result_KDE$RMSE[i_simu_setting] = sqrt(mean((intensity_pred_best - intensity_values)^2))

    # 2.5 BART
    # --------------------
    save_file_name = sprintf('%s/BART_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    df_result_BART$RMSE[i_simu_setting] = sqrt(mean((intensity_pred - intensity_values)^2))
    
}
```

```r
## Shot intensity

source("../RealDataAnalysis/Functions/basket_func_same.R")
z = c(1,0,1); type = 1
logintensity_function <- function(x, y, z, type, theta_alpha, theta_beta_type0, theta_beta_type1, n, a, b) {

        coords = matrix(0, length(x), 2)
        coords[,1] = x;  coords[,2] = y 
        GP_basis_value = GP.eigen.funcs(coords, n=n, a=a, b=b) 

        if(type == 0) logintensity = GP_basis_value%*%theta_alpha + GP_basis_value %*% theta_beta_type0 %*% z
        if(type == 1) logintensity = GP_basis_value%*%theta_alpha + GP_basis_value %*% theta_beta_type1 %*% z

      return( as.vector(logintensity) ) 
}
intensity_function <- function(...) {
    return(exp(logintensity_function(...)))
}

df = data.frame(coord1 = 1, coord2 = 1, intensity_pred = 1, intensity = 1, setting = '1', type = 1, home = 1, level = 1)
for(setting_string in c('simu','curry')){for(type in c(0,1)){for(home in c(0,1)){for(level in c(0,1)){
    z = c(home, 1-home, level)
    i_rep = 1
    save_subfolder = sprintf('%s/%s', save_folder, setting_string)
    save_file_name = sprintf('%s/DATA_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)
    save_file_name = sprintf('%s/SVCLGCP_rep=%d.RData', save_subfolder, i_rep)
    load(file = save_file_name)

    p = dim(covar)[2] - 1
    intergrid = as.matrix(expand.grid(seq(-1,1,length=60),seq(-1,1,length=60)))
    theta_alpha = theta[,1,drop = FALSE]
    theta_beta_type0 = theta[,2:(1+p),drop = FALSE]
    theta_beta_type1 = theta[,(2+p):(1+2*p),drop = FALSE]


    intensity_values_on_grid = intensity_function(x = intergrid[,1], y = intergrid[,2],
                                  z = z, type = type, 
                                  theta_alpha = theta_alpha, theta_beta_type0 = theta_beta_type0, theta_beta_type1 = theta_beta_type1,
                                  n = n, a = a, b = b)
    
    theta_hat = apply(fit$Theta.result[simu_info$num_burn+1:simu_info$num_save,,],c(2,3),mean)
    intensity_pred_on_grid = intensity_function(x = intergrid[,1], y = intergrid[,2],
                                    z = z, type = type, 
                                    theta_alpha = theta_hat[,1],
                                    theta_beta_type0 = theta_hat[,2:(p+1)],
                                    theta_beta_type1 = theta_hat[,(p+2):(2*p+1)],
                                    n = n, a = a, b = b)
    
    
    df = rbind(df , data.frame(coord1 = 23.5*(intergrid[,2]+1), coord2 = 25*intergrid[,1], 
                               intensity = intensity_values_on_grid, intensity_pred = intensity_pred_on_grid, 
                               setting = setting_string, type = type, home = home, level = level) )  
}}}}
df = df[-1, ]
df$setting = factor(df$setting, levels = c('simu','curry')); levels(df$setting) = c('Synthetic Intensity','Curry Intensity')


# plot intensity
###################
num_intervals = 10
digit_kept = 2

home_selected = 1; level_selected = 1; type_selected = 1
df_intensity = df %>% filter(home == home_selected , level == level_selected , type == type_selected)
df_intensity_pred = df_intensity; df_intensity_pred$intensity = df_intensity_pred$intensity_pred
df_intensity = rbind(df_intensity, df_intensity_pred)
df_intensity$method = rep(c('Truth','Estimated'), each = dim(df_intensity_pred)[1])

value_of_interest = df_intensity$intensity
value_quantiles = quantile(value_of_interest, probs = seq(0, num_intervals)/num_intervals)
breaks = round(value_quantiles, digit_kept)
breaks[1] = floor(value_quantiles[1]*10^digit_kept)/10^digit_kept;  breaks[num_intervals+1] = ceiling(value_quantiles[num_intervals+1]*10^digit_kept)/10^digit_kept
breaks = unique(breaks)
p_intensity = ggshotintensity(intensity = df_intensity, x = "coord1", y = "coord2", value_col = "intensity", breaks = breaks, label_name = 'intensity     ') +
      geom_tile() +
      facet_grid(rows = vars(method), cols = vars(setting)) + 
      theme(axis.ticks = element_blank(), axis.text = element_blank(), legend.title = element_text(size=25),legend.text = element_text(size=15), legend.key.size = unit(.8, 'cm')) +
      theme(strip.text=element_text(size=30), panel.spacing = unit(1, 'lines'), legend.position = "left") 


# plot RR
###################
df_prob = df %>% filter(level == 1) %>% group_by(setting, home, level) %>% mutate( prob = intensity / sum(intensity), prob_pred = intensity_pred / sum(intensity_pred) ) # sum over type
df_RR = df_prob %>% filter(home == 1, type == 1) #%>% mutate(RR = prob / (df_prob %>% filter(home == 1, type == 1))$prob )
df_RR$RR = (df_prob %>% filter(home == 1, type == 1))$prob / (df_prob %>% filter(home == 0, type == 1))$prob
df_RR$RR_pred = (df_prob %>% filter(home == 1, type == 1))$prob_pred / (df_prob %>% filter(home == 0, type == 1))$prob_pred

df_RR = df_RR
df_RR_pred = df_RR; df_RR_pred$RR = df_RR$RR_pred
df_RR = rbind(df_RR, df_RR_pred)
df_RR$method = rep(c('Truth','Estimated'), each = dim(df_RR_pred)[1])
levels(df_RR$setting) = c('Synthetic Risk','Curry Risk')

value_of_interest = df_RR$RR
value_quantiles = quantile(value_of_interest, probs = seq(0, num_intervals)/num_intervals)
breaks = round(value_quantiles, digit_kept)
breaks[1] = floor(value_quantiles[1]*10^digit_kept)/10^digit_kept;  breaks[num_intervals+1] = ceiling(value_quantiles[num_intervals+1]*10^digit_kept)/10^digit_kept
breaks = unique(breaks)

p_RR = ggshotintensity(intensity = df_RR, x = "coord1", y = "coord2", value_col = "RR", breaks = breaks, label_name = 'relative risk') + 
      geom_tile() + 
      facet_grid(rows = vars(method), cols = vars(setting)) + 
      theme(axis.ticks = element_blank(), axis.text = element_blank(), legend.title = element_text(size=25),legend.text = element_text(size=15), legend.key.size = unit(.8, 'cm')) +
      theme(strip.text=element_text(size=30), panel.spacing = unit(1, 'lines'), legend.position = "right") 



p_summary = grid.arrange(p_intensity, p_RR, ncol = 2)
ggsave(p_summary ,filename = sprintf('%s/Summary.pdf', save_folder), width = 20, height = 10)
```
